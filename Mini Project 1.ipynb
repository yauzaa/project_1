Which month (arrival_date_month) had the most cancellations for bookings of City Hotel type in 2015? 2016? 2017?:

```python
city_hotel_bookings = bookings.query("hotel == 'City Hotel'")

grouped_data = city_hotel_bookings.groupby(['arrival_date_year', 'arrival_date_month'])

cancelled_counts = grouped_data['is_canceled'].sum()

for year in [2015, 2016, 2017]:
    max_cancelled_month = cancelled_counts.loc[year].idxmax()
    max_cancelled_count = cancelled_counts.loc[year, max_cancelled_month]
    print(f"In {year}, the month with the most cancellations for City Hotel bookings was {max_cancelled_month} with {max_cancelled_count} cancellations.")
```
